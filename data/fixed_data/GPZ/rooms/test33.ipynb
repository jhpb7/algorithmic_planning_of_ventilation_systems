```python
import pyomo.environ as pyo

m = pyo.AbstractModel()

m.x = pyo.Set(dimen=2)

m.vf = pyo.Param(m.x)

def splitter_block(block, flag_ac, i,j):
    vs_scenario = block.parent_block()
    
    block.volume_flow = pyo.Param(initialize=1)#vs_scenario.vf[i,j]
    return block

flag_ac = "tree"

m.b = pyo.Block(m.x, rule = lambda m, i, j: splitter_block(m, flag_ac, i,j))
```
